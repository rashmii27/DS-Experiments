**Encoding techniques**

```python
#Label Encoding
label_encoder = LabelEncoder()
data['Grade_labelencoded'] = label_encoder.fit_transform(data['Grade'])
print(data)
```

```python
#one-hot encoding
Grade_onehotencoded = pd.get_dummies(data, columns=['Grade','Name'])
```

```python
#Binary encoding
import category_encoders as ce


binary_encoder = ce.BinaryEncoder(cols=['Grade'])
binary_encoded = binary_encoder.fit_transform(data['Grade'])


binary_encoded = pd.concat([data, binary_encoded], axis=1)


print(binary_encoded.head())
```

```python
#frequency encoding
frequency_encoding = data['Grade'].value_counts().to_dict()
data['frequency_encoded'] = data['Grade'].map(frequency_encoding)


print(data.head())
```
